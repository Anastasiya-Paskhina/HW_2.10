```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
folder = 'names'
directory = ''.join((os.getcwd(), '\\', folder, '\\'))
dir_list = os.listdir(path=directory)
```

```python
dir_list = {}
for year in range(1900, 2001):
    dir_list[year] = pd.read_csv(f'names/yob{year}.txt', names = ['Name', 'Gender', 'Count'])
all_names = pd.concat(dir_list).reset_index(level=1, drop=True)
```

1. Построить график изменения количества имен Ruth и Robert с 1900 по 2000. 

```python
dyn_names = all_names[(all_names['Name'].isin(['Ruth', 'Robert']))]
dyn_names = dyn_names.groupby(['Name', dyn_names.index]).sum()
dyn_names.unstack('Name').plot()
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.grid()
```

```python
dyn_names_2 = all_names[(all_names['Name'].isin(['Ruth', 'Robert']))]
dyn_names_2 = dyn_names_2[dyn_names_2.index.isin(range(1900, 2001, 5))]
dyn_names_2 = dyn_names_2.groupby(['Name', dyn_names_2.index]).sum()
dyn_names_2.plot(kind='hist')
```

```python
dyn_names_3 = all_names[(all_names['Name'].str[0] == 'R')]
dyn_names_3 = dyn_names_3[dyn_names_3.index == 1950]
dyn_names_3 = dyn_names_3.groupby(['Name']).sum().sort_values(['Count'], ascending=False).head(10)   
dyn_names_3.plot(kind='pie', y='Count', legend=False)
```

```python
dyn_names_4 = all_names.groupby(['Name']).sum()
dyn_names_4['Letters'] = sum(dyn_names_4.index.str.count(letter) for letter in 'AEIOUYaeiouy')
dyn_names_4.plot(kind='scatter', x='Letters', y='Count', c='green')
```
